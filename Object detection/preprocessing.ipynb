```
import os
```

```
def reconstruct(txt, lines):
    re = []

    for line in lines:
        ele = line.split(" ")[1:]
        ele[3] = str(ele[3]).split("\n")[0]
        if ele[0] == ele[2] or ele[1] == ele[3]:
            continue
        ele = txt + "," + ele[0] +  "," + ele[1] + "," + ele[2] + "," + str(ele[3]) + ",word"
        re.append(ele)

    return re
```

```
def writeline(txt,re):
    path = "Colab Notebooks/RetinaNet/data3_label/" + txt
    with(open(path, 'a')) as f:
        for r in re:       
            f.writelines(r + "\n")
```

```
for txt in os.listdir("TKHMTH/TKH/label_char"):
    path = "TKHMTH/TKH/label_char/" + txt
    with(open(path, 'r')) as f:
        lines = f.readlines()
    jpg = txt.split(".")[0] + ".jpg"
    re = reconstruct(jpg, lines)
    writeline(txt,re)
```
